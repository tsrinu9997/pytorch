# PyTorch Functional API: SimpleNN Explanation

### 1. Class Definition and Inheritance
`class SimpleNN(nn.Module):`
*   **Inheritance:** Your network must inherit from `nn.Module`, the base class for all neural network modules in PyTorch.
*   **Purpose:** This inheritance allows PyTorch to automatically track and update the "learnable" weights and biases within your layers.

---

### 2. The Initialization Method (`__init__`)
This is where you define the structural **building blocks** of your network.

*   `super(SimpleNN, self).__init__()`: This line is required to properly initialize the parent `nn.Module` class.
*   `self.fc1 = nn.Linear(input_size, hidden_size)`: Defines the first **Linear layer** (fully connected). It transforms data from the input dimension to a hidden dimension.
*   `self.relu = nn.ReLU()`: Sets up the **ReLU activation function**. This introduces non-linearity, allowing the network to learn complex patterns.
*   `self.fc2 = nn.Linear(hidden_size, output_size)`: Defines the second linear layer, which transforms the hidden data into the final output size (e.g., number of classes).

---

### 3. The Forward Pass (`forward`)
This method defines the actual path data takes as it flows through the network.

1.  `x = self.fc1(x)`: Pass input `x` through the first linear layer.
2.  `x = self.relu(x)`: Apply the ReLU activation to the output.
3.  `x = self.fc2(x)`: Pass the result through the final output layer.
4.  **Return:** `return x` returns the final prediction or scores.


````python
## Functional API

class SimpleNN(nn.Module):

  def __init__(self, input_size,hidden_size, output_size):

    super(SimpleNN, self).__init__()
    self.fc1 = nn.Linear(input_size, hidden_size)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(hidden_size, output_size)

  def forward(self, x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
````

# PyTorch Functional API: SimpleNN Explanation

### 1. Class Definition and Inheritance
`class SimpleNN(nn.Module):`
*   **Inheritance:** Your network must inherit from `nn.Module`, the base class for all neural network modules in PyTorch.
*   **Purpose:** This inheritance allows PyTorch to automatically track and update the "learnable" weights and biases within your layers.

---

### 2. The Initialization Method (`__init__`)
This is the setup phase where you create the structural **building blocks** of your network.

*   `super(SimpleNN, self).__init__()`: Required to properly initialize the parent `nn.Module` class so it can track your layers.
*   `self.fc1 = nn.Linear(input_size, hidden_size)`: The **input layer**. It transforms raw data into a larger hidden space (hidden neurons).
*   `self.relu = nn.ReLU()`: The **activation function**. It introduces non-linearity, allowing the network to learn complex, "curvy" patterns.
*   `self.fc2 = nn.Linear(hidden_size, output_size)`: The **output layer**. It condenses the hidden information into your final prediction count.

---

### 3. The Forward Pass (`forward`)
This method defines the **assembly line**—the actual path data takes as it flows through the network.

1.  `x = self.fc1(x)`: The raw data `x` is processed by the first linear layer.
2.  `x = self.relu(x)`: The result is filtered through the ReLU activation.
3.  `x = self.fc2(x)`: The filtered data is processed by the final layer to get the final result.
4.  **Return:** `return x` sends back the model's final prediction or "guess."


````python
## Sequential

class SimpleNNSequential(nn.Module):

  def __init__(self, input_size,hidden_size, output_size):
    super(SimpleNN, self).__init__()

    self.network = nn.Sequential(
        nn.Linear(input_size, hidden_size),
        nn.ReLU(),
        nn.Linear(hidden_size, output_size),
    )

  def forward(self, x):
    return self.network(x)
````

## Training the neural network

### 4. Model Initialization and Inspection
To use the network, we create an instance and define its dimensions.

*   **`model_func = SimpleNN(input_size=4, hidden_size=8, output_size=3)`**: Creates the model instance. 
    *   `input_size=4`: The model expects 4 input features.
    *   `hidden_size=8`: The model uses 8 neurons in the hidden layer.
    *   `output_size=3`: The model will output 3 values (scores).
*   **`print(model_func)`**: Prints the architecture summary to the console so you can verify the layer structure.


````python
model_func = SimpleNN(input_size=4, hidden_size=8, output_size=3)
print(model_func)
````

### 5. Data Preparation and Training Setup
Before training, we define our data, how we measure error, and how we update the model.

*   **`X = torch.randn(10, 4)`**: Creates a batch of **10 input samples**, each having **4 features**.
*   **`Y = torch.randint(0, 3, (10,))`**: Creates **10 random ground-truth labels** (classes 0, 1, or 2).
*   **`criterion = nn.CrossEntropyLoss()`**: Sets the **Loss Function**. This calculates the difference between the model's prediction and the actual label.
*   **`optimizer = optim.Adam(model_func.parameters(), lr=0.01)`**: Initializes the **Optimizer**. 
    *   `model_func.parameters()`: Connects the optimizer to the model's weights.
    *   `lr=0.01`: Sets the **Learning Rate**, controlling the speed of the training process.


````python
X = torch.randn(10, 4) # 10 samples, 4 feature
Y = torch.randint(0, 3, (10,))

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_func.parameters(), lr=0.01)
````

### 6. The Training Loop
This is the core iterative process where the model learns and improves its predictions.

*   **`epoch = 120`**: Defines the number of complete passes through the dataset.
*   **`optimizer.zero_grad()`**: **Clears the gradients**. This ensures we start with a clean slate for each training step.
*   **`outputs = model_func(X)`**: Runs the **Forward Pass** to get predictions.
*   **`loss = criterion(outputs, Y)`**: Computes the **Loss** to see how far off the predictions are.
*   **`loss.backward()`**: Runs **Backpropagation** to find the gradients for all weights.
*   **`optimizer.step()`**: Updates the **Model Weights** to minimize the error.
*   **`if (e+1) % 10 == 0`**: Monitors progress by printing the current loss every **10 epochs**.


````python
# training loop

epoch = 120

for e in range(epoch):
  optimizer.zero_grad()
  outputs = model_func(X)
  loss = criterion(outputs, Y)
  loss.backward()
  optimizer.step()

  if (e+1) % 10 == 0:
    print(f"Epoch [{e+1}]/50, Loss : {loss.item() :.4f}")
````
