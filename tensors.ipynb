## Tensor Data Types

| **Data Type**             | **Dtype**         | **Description**                                                                                                                                                                |
|---------------------------|-------------------|--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **32-bit Floating Point** | `torch.float32`   | Standard floating-point type used for most deep learning tasks. Provides a balance between precision and memory usage.                                                         |
| **64-bit Floating Point** | `torch.float64`   | Double-precision floating point. Useful for high-precision numerical tasks but uses more memory.                                                                               |
| **16-bit Floating Point** | `torch.float16`   | Half-precision floating point. Commonly used in mixed-precision training to reduce memory and computational overhead on modern GPUs.                                            |
| **BFloat16**              | `torch.bfloat16`  | Brain floating-point format with reduced precision compared to `float16`. Used in mixed-precision training, especially on TPUs.                                                |
| **8-bit Floating Point**  | `torch.float8`    | Ultra-low-precision floating point. Used for experimental applications and extreme memory-constrained environments (less common).                                               |
| **8-bit Integer**         | `torch.int8`      | 8-bit signed integer. Used for quantized models to save memory and computation in inference.                                                                                   |
| **16-bit Integer**        | `torch.int16`     | 16-bit signed integer. Useful for special numerical tasks requiring intermediate precision.                                                                                    |
| **32-bit Integer**        | `torch.int32`     | Standard signed integer type. Commonly used for indexing and general-purpose numerical tasks.                                                                                  |
| **64-bit Integer**        | `torch.int64`     | Long integer type. Often used for large indexing arrays or for tasks involving large numbers.                                                                                  |
| **8-bit Unsigned Integer**| `torch.uint8`     | 8-bit unsigned integer. Commonly used for image data (e.g., pixel values between 0 and 255).                                                                                    |
| **Boolean**               | `torch.bool`      | Boolean type, stores `True` or `False` values. Often used for masks in logical operations.                                                                                      |
| **Complex 64**            | `torch.complex64` | Complex number type with 32-bit real and 32-bit imaginary parts. Used for scientific and signal processing tasks.                                                               |
| **Complex 128**           | `torch.complex128`| Complex number type with 64-bit real and 64-bit imaginary parts. Offers higher precision but uses more memory.                                                                 |
| **Quantized Integer**     | `torch.qint8`     | Quantized signed 8-bit integer. Used in quantized models for efficient inference.                                                                                              |
| **Quantized Unsigned Integer** | `torch.quint8` | Quantized unsigned 8-bit integer. Often used for quantized tensors in image-related tasks.                                                                                     |


### Data types for tensors


- floating -- for most deep learning tasks ---> float32, float64, float16
- integers - for categorical data and indices --> int32, int64, int8

### Transposing and premuting
```
These operations allow you to rearrange the dimensions of a tensor, which is critical for matching data formats (like switching between Image-First and Channel-First layouts).
```

- transpose() -->swaps two dimesnion. mxn --> nxm
- premute() --> rearranges all dimension in the specified order


| Feature | `transpose` | `permute` |
| :--- | :--- | :--- |
| **Scope** | Swaps only 2 dimensions. | Rearranges any number of dimensions. |
| **Complexity** | Simple, specific swaps. | Flexible, multi-axis reordering. |
| **Common Use** | Matrix math / Linear algebra. | Preprocessing image or video batches. |
| **Example** | `x.transpose(0, 1)` | `x.permute(2, 0, 1)` |


````python
# transpose

tensor_original = torch.arange(24).reshape(12,2)

transposed_tensor = tensor_original.transpose(0,1)

print(tensor_original)

print(tensor_original.shape)
print(transposed_tensor.shape)

print(transposed_tensor)
````

````python
tensor_original = torch.arange(24).reshape(2,3,4)

print(2*3*4)

print(tensor_original.shape)
print(tensor_original)

# [0,1,2]
permuted_tensor = tensor_original.permute(2, 0, 1)
print(permuted_tensor)
print(permuted_tensor.shape)
````

````python

# 1. Create a sample image tensor [Height, Width, Channels]
# Shape: (224, 224, 3)
image_tensor = torch.randn(224, 224, 3) 

# 2. Using transpose()
# Swap Height (0) and Width (1) -> Shape: (224, 224, 3)
# Useful for rotating or flipping specific dimensions
transposed = image_tensor.transpose(0, 1)

# 3. Using permute() 
# Reorder to [Channels, Height, Width] -> Shape: (3, 224, 224)
# Standard format for PyTorch models (CNNs)
permuted = image_tensor.permute(2, 0, 1)

print(f"Original:   {image_tensor.shape}")
print(f"Transpose:  {transposed.shape}")
print(f"Permute:    {permuted.shape}")
````

````python
# 1. Create a contiguous tensor
x = torch.randn(3, 2)
print(x.is_contiguous()) # Output: True

# 2. Transpose it (becomes non-contiguous)
y = x.transpose(0, 1)
print(y.is_contiguous()) # Output: False

# 3. This will CRASH with a RuntimeError
# try:
#     z = y.view(-1) 
# except RuntimeError as e:
#     print(f"Error: {e}")

# 4. Use .contiguous() to fix it
y_fixed = y.contiguous()
print(y_fixed.is_contiguous()) # Output: True

# Now .view() works perfectly
z = y_fixed.view(-1)
````

### Cloning and detaching


| Method | New Memory? | Tracks Gradients? | Common Use Case |
| :--- | :---: | :---: | :--- |
| **`.clone()`** | Yes | **Yes** | Making a copy to modify without changing the original's values, but keeping the math link. |
| **`.detach()`** | No (View) | **No** | Moving a tensor to a part of the code where you don't want backpropagation (like logging). |
| **`.detach().clone()`** | Yes | **No** | Creating a safe, independent copy for external libraries like Matplotlib or NumPy. |


````python
import torch

# Original tensor with gradient tracking enabled
original = torch.tensor([1.0, 2.0], requires_grad=True)

# 1. CLONE: Same data, still tracks gradients
cloned = original.clone() 
# cloned.requires_grad is True

# 2. DETACH: Same data, STOP tracking gradients
detached = original.detach() 
# detached.requires_grad is False

# 3. BOTH (Very Common): Clean copy for numpy or plotting
# This gives you a separate copy that won't affect the model training
final_copy = original.detach().clone()
````

````python
import copy

a = [1,2,3]
b = copy.deepcopy(a)

print(a)
print(b)
````

````python
tensor = torch.ones(3,3, requires_grad=True) # part of computation graph

cloned_tensor = tensor.clone() # # part of computation graph
print(cloned_tensor)
detached_tensor = tensor.detach() # this code will detach the tensor from the computation graph
# # non a part of computation graph
# but storage will be same as the original ones
print(detached_tensor)
````

### 3. Reduction operation



| Operation | Dimension | Result Shape | Description |
| :--- | :--- | :--- | :--- |
| `torch.sum(f)` | None | Scalar | **Total** of all elements |
| `torch.sum(f, dim=0)` | 0 (Rows) | `(3,)` | Sum of each **column** |
| `torch.sum(f, dim=1)` | 1 (Cols) | `(2,)` | Sum of each **row** |


````python
# This code creates a tensor filled with random integers between 0 and 9, but stores them as 32-bit floating-point numbers. 

f = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
f
````

````python
# sum
# dim=0: Vertical axis (columns). Summing here collapses rows into a single row.
# dim=1: Horizontal axis (rows). Summing here collapses columns into a single column. 
print(torch.sum(f))
# sum along columns
print(torch.sum(f, dim=0))
# sum along rows
print(torch.sum(f, dim=1))
````
